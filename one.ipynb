Added pressure sensor implementation and data processing

```python
import numpy as np
import pandas as pd

Vcc = 5.0
R_fixed = 10000  # 10kΩ

def pressure_voltage(pressure):
    # pressure in Newtons (0–100)
    R_min = 200     # high pressure
    R_max = 10000  # no pressure

    # inverse relationship
    R_sensor = R_max - (R_max - R_min) * (pressure / 100)
    V_out = Vcc * (R_fixed / (R_fixed + R_sensor))
    return V_out


data = []

for _ in range(300):  # total samples

    gesture = np.random.choice([0, 1, 2])

    if gesture == 0:  # NO CONTACT
        pressures = np.random.uniform(0, 5, 4)

    elif gesture == 1:  # GRASP
        pressures = np.random.uniform(40, 80, 4)

    else:  # PRESS OBJECT
        pressures = [
            np.random.uniform(60, 90),  # thumb
            np.random.uniform(60, 90),  # index
            np.random.uniform(20, 40),  # middle
            np.random.uniform(70, 100)  # thenar
        ]

    voltages = [pressure_voltage(p) for p in pressures]
    data.append(voltages + [gesture])


columns = ["Thumb_P", "Index_P", "Middle_P", "Thenar_P", "Gesture"]
df = pd.DataFrame(data, columns=columns)

df.to_csv("pressure_sensor_dataset.csv", index=False)
print("Dataset saved as pressure_sensor_dataset.csv")

df.head()
```
